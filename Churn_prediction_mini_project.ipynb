```python
#dropping null values
df.dropna()
```

```python
#mapping
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0}).dropna().astype(int)
df['multi_screen'] = df['multi_screen'].map({'yes': 1, 'no': 0}).dropna().astype(int)
df['mail_subscribed'] = df['mail_subscribed'].map({'yes': 1, 'no': 0}).dropna().astype(int) 
```

```python
#convert to numeric values
df['churn'] = pd.to_numeric(df['churn'], errors='coerce').dropna().astype(int)
```

```python
#Filling missing NaN values
df['gender'] = df['gender'].fillna(df['gender'].median())
df['multi_screen'] = df['multi_screen'].fillna(df['multi_screen'].median())
df['mail_subscribed'] = df['mail_subscribed'].fillna(df['mail_subscribed'].median())
df['churn'] = df['churn'].fillna(df['churn'].median())
```

```python
#checking NaN values 
print(df['gender'].isna().sum())
print(df['multi_screen'].isna().sum())
print(df['mail_subscribed'].isna().sum())
print(df['churn'].isna().sum())
```

```python
df.head(10)
```

```python
# Check for other columns with object dtype and convert them to numeric if necessary

for col in df.select_dtypes(include=['object']).columns:
    # If column is not 'phone_no' or 'customer_id', convert to numeric
    if col not in ['phone_no', 'customer_id']:
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(-1).astype(int)
        except ValueError:
            print(f"Column '{col}' could not be converted to numeric. Please inspect its values.")
```

FEATURE SELECTION

```python
#Feature selection
X= df.drop(columns= ['phone_no', 'customer_id', 'year','churn'])
y= df['churn']
```

```python
y.head()
```

DATA SPLITTING

```python
#splitting of data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print("Unique classes in y_test:", set(y_test))
```

MODEL

```python
from sklearn.tree import DecisionTreeClassifier

t= DecisionTreeClassifier()
t
```

```python
t.fit(X_train, y_train)
```

```python
y_predict = t.predict(X_test)
```

EVALUATION METRICS

CLASSIFICATION REPORT 

```python
#Classification report 
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import pandas as pd

cr= print(classification_report(y_test, y_predict))
```

```python
#converting classification report as a dictionary
cr_dict = classification_report(y_test, y_predict, output_dict=True)

# Convert the dictionary to a DataFrame
cr_df = pd.DataFrame(cr_dict).transpose()

# Drop support column for plot
metrics_to_plot = cr_df.drop(columns=['support'])

# Plot classification report
plt.figure(figsize=(10, 6))
sns.heatmap(metrics_to_plot, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Classification Report')
plt.xlabel('Metrics')
plt.ylabel('Classes')
plt.show()
```

```python
#Confusion metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm= confusion_matrix(y_test, y_predict)
cm
```

```python

import matplotlib.pyplot as plt

# Predict with the best model
y_predict = t.predict(X_test)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=t.classes_)
disp.plot(cmap=plt.cm.Purples)

plt.title("Confusion Matrix - Decision Tree")
plt.grid(False)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt


#Get predicted probabilities for class 1 (churn = 1)
y_probability = t.predict_proba(X_test)[:, 1]

def plot_roc_curve(y_test, y_probability, t):
    fpr, tpr, thresholds = roc_curve(y_test, y_probability)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='red', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC-AUC Curve for {t}')
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

# Call the function
plot_roc_curve(y_test, y_probability, 'Decision Tree')
```
